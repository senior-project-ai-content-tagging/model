# Text Preprocessing

## Clean Data

```python
import re

def clean_data(text):
  text = re.sub('[^\u0E00-\u0E7F\s\.]', '', text)
  text = re.sub('\B\.\B','',text)
  text = re.sub('\s+', ' ', text)
  return text
```

```python
clean_data(df.contents[0])
```

## Tokenization

```python
from pythainlp import word_tokenize
from pythainlp.util import normalize

def tokenize(text):
  return word_tokenize(text, keep_whitespace=False) 
```

## Stop word removal

```python
from pythainlp.corpus.common import thai_stopwords
stopwords = list(thai_stopwords())
```

```python
def remove_stopword(tokens):
  token = [w for w in tokens if w.lower not in stopwords and w != '.']
  text = ' '.join(word for word in token)
  return text
```

## Final Preprocessing

```python
def preprocess(text):
  # clean data
  text = clean_data(text)
  # normalize
  text = normalize(text)
  # tokenize 
  token = tokenize(text)
  # remove stop words from tokens
  text = remove_stopword(token)
  return text
```

```python
df['clean'] = df['contents'].apply(preprocess)
df.head()
```

# TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(
  tokenizer = lambda x:x.split(' '), # เพิ่ม Analyzer สำหรับตัดคำ
  stop_words = stopwords,            # ป้อนรายการคำ Stop words ที่เตรียมไว้
  ngram_range = (1,1),               
  min_df = 5,                        # ขั้นต่ำของ Doc Freq ของ Term
  max_df = 0.95,
  max_features = int(unique_word*0.5),
  norm='l2'
)
```

# Prepare Train & Test Dataset

```python
from sklearn.model_selection import train_test_split
```

```python
X = df['clean']
y = df['tag']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101, stratify=df['tag'])
```

```python
X_train = tfidf.fit_transform(X_train)
X_test = tfidf.transform(X_test)
```

# KNN

```python
import optuna
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

def objective(trial):
    n_neighbors = trial.suggest_int('n_neighbors', 1, 30)
    weights = trial.suggest_categorical('weights', ['uniform', 'distance'])
    knn = KNeighborsClassifier(n_neighbors = n_neighbors, 
                           metric='euclidean',
                           n_jobs=-1,
                           weights=weights)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    return accuracy_score(y_test, y_pred)
```

```python
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=50)
print('Best hyperparameters:', study.best_params)
```

```python
best_params = study.best_params
knn = KNeighborsClassifier(n_neighbors = best_params['n_neighbors'], 
                           metric='euclidean',
                           n_jobs=-1,
                           weights=best_params['weights'])
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
print('Accuracy:', accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
